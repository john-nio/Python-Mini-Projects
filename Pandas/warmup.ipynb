## Learning Query

```python
pokemon_data = pd.read_csv("Data/Pokemon.csv", index_col="Name")

pokemon = input ("Enter the name of the Pokemon")

try:
    print(pokemon_data.loc[pokemon])
except KeyError:
    print(f" {pokemon} is not found")
```

```python
pokemon_data.head
```

## Finding Water Pokemon

```python
water_pokemon = pokemon_data[(pokemon_data['Type1']== 'Water') | (pokemon_data['Type2']== 'Water')]

print(water_pokemon)
```
